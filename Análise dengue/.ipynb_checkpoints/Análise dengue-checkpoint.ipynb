```python
import pandas as pd

dengue = pd.read_csv('dengue-dataset.csv', sep=',', encoding='utf-8')
```

```python
dengue.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def time_stamp(begin, end, column):
    dates = []
    values = []
    
    f_begin = False
    
    for index, row in dengue.iterrows():
        if (row["data"] == begin):
            f_begin = True
        
        if (f_begin):
            dates.append(row["data"])
            values.append(row[column])
        
        if (row["data"] == end):
            f_begin = False
            
    
    return (dates, values)
    
```

```python
tr = time_stamp('1998-01-01', '1998-12-01', 'casos-confirmados')
```

```python
#plot simples

fig, ax = plt.subplots()
plt.plot_date(tr[0], tr[1])
ax.xaxis.set_tick_params(rotation=60, labelsize=10)

plt.show()
```

```python
ts = time_stamp('1998-01-01', '1998-12-01', 'chuva')
```

```python
#plot simples

fig, ax = plt.subplots()
plt.plot_date(ts[0], ts[1])
ax.xaxis.set_tick_params(rotation=60, labelsize=10)

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,3))
fig.suptitle('Os dois gráficos')

axs[0].xaxis.set_tick_params(rotation=60, labelsize=10)
axs[0].set_title('Casos de dengue')
axs[0].plot(tr[0], tr[1])

axs[1].xaxis.set_tick_params(rotation=60, labelsize=10)
axs[1].set_title('Litros de chuva')
axs[1].plot(ts[0], ts[1])

plt.show()
```

```python
import matplotlib.dates as mdates

data = pd.read_csv('dengue-dataset.csv', usecols=['data','chuva'], parse_dates=['data'])
data.set_index('data',inplace=True)

fig, ax = plt.subplots(figsize=(24,7))
data.plot(ax=ax)


ax.xaxis.set_major_locator(mdates.WeekdayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
```
