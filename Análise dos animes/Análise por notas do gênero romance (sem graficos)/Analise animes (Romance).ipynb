```python
# Primeiro, importamos as bibliotecas
import pandas as pd
```

```python
# Agora, lemos o dataset
Dados_Animes = pd.read_csv("DadosAnime.csv")
# Exibimos a variável para conferir se a leitura foi realizada de forma correta
Dados_Animes
```

```python
# Para realizar a busca por gêneros, 
# precisamos percorrer a coluna genre e verificar a string que guarda os gêneros dos animes
index = []
count = 0
for i in Dados_Animes['genre']:
    j = i.split(',')
    for k in j:
        if(k.find('Romance') != -1):
            index.append(count)
    count+=1
index
```

```python
# Agora, pegamos todas as notas e guardamos numa lista de notas
Notas = Dados_Animes['rating'][index]
Notas
```

```python
# Aqui, faremos as mesmas coisas que fizemos com as notas, mas para guardar os nomes dos animes
Nomes = Dados_Animes['name'][index]
Nomes
```

```python
# Com os dados em mãos, vamos criar uma lista apenas com as notas e nomes dos animes
Analise_Romance = pd.concat([Nomes, Notas], axis=1)
Analise_Romance
```

```python
# Agora, ordenamos a lista utilizando a coluna de notas como base
Analise_Romance.sort_values(by='rating')
Analise_Romance
```
